```python
# Create a function (OddEven) that checks whether the given number is Odd or Even
```

```python
def OddEven():
    u_input=int(input("Enter a number:"))
    if(u_input%2==0):
        print(u_input, "is Even Number")
    else:
        print(u_input, "is Odd Number")
```

```python
OddEven()
```

```python
# Create a function (Eligible) that tells eligibility of marriage for male and female according to their age limit like 21 for male and 18 for femal
```

```python
def Eligible(Gender, Age):
    if(Gender=="Male"):
        print("Your Gender:", Gender)
        print("Your Age:", Age)
        if(Age>=21):
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")
    elif(Gender=="Female"):
        print("Your Gender:", Gender)
        print("Your Age:", Age)
        if(Age>=18):
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")
```

```python
Eligible("Male", 20)
```
