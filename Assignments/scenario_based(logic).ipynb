```python
def largest_number():
    num_list=[34,23,67,12,98,45]
    v_larg_num=num_list[0]
    for i in num_list:
        if(i>v_larg_num):
            v_larg_num=i
    print(v_larg_num)
```

```python
# 5. Scenario: A text-processing tool reverses a given word or sentence for formatting purposes.
# Write logic to take a word or sentence as input and produce its reversed version.
```

```python
input_str=input("Enter a string:")
v_len=len(input_str)
print("Your reversed string is:")
for i in range(v_len-1,-1,-1):
    rev_str=input_str[i]
    print(rev_str)
```

```python
# 6. Scenario: A grading system determines whether a student has passed or failed based on their score.
# Write logic to check if a student has passed a subject by scoring at least 40 marks.
```

```python
def PassFailScore(v_subj_mark_1, v_subj_mark_2, v_subj_mark_3, v_subj_mark_4, v_subj_mark_5, v_subj_mark_6):
    Subj_list=['Maths', 'English', 'Physics', 'Chemistry', 'Computer Science', 'Language1']
    Marks_list=[v_subj_mark_1, v_subj_mark_2, v_subj_mark_3, v_subj_mark_4, v_subj_mark_5, v_subj_mark_6]
    #print("Enter your marks for the following Subjects in sequence: Maths, English, Physics, Chemistry, Computer Science, Language1:")
    total_score=((v_subj_mark_1+v_subj_mark_2+v_subj_mark_3+v_subj_mark_4+v_subj_mark_5+v_subj_mark_6)/600)*100
    for i in range(len(Subj_list)):
        if(Marks_list[i]>=40):
            print("You have passed in ", Subj_list[i], "subject with", Marks_list[i], "marks")
        else:
            print("Sorry, you have failed the exam with" , Marks_list[i], "marks in", Subj_list[i], "subject - You can still retry in your subsequent attempts")
    print("You total score is:",total_score)
    return Marks_list

def OverAllResult(Marks_list):
    for i in Marks_list:
        #print(i)
        if(i<40):
            print("You overall exam result is: Fail")
            return
    print("Your overall exam result is: Pass")          
```

```python
Result_Stud_1=PassFailScore(40,56,56,90,37,89)
```

```python
OverAllResult(Result_Stud_1)
```

```python
Result_Stud_2=PassFailScore(89,41,56,90,67,92)
```

```python
OverAllResult(Result_Stud_2)
```

```python
# 7. Scenario: A retail store offers a 20% discount if a customer’s total order exceeds $100. 
# Write logic to calculate the final amount to be paid after applying the discount.
```

```python
final_amt=0
def discount_amount():
    total_order_amount=float(input("Enter the total order value: "))
    if(total_order_amount>100):
        discount_offer=(20/100)
        discount_value=total_order_amount*discount_offer
        final_amount=total_order_amount-discount_value
        print("Discount applied: $",discount_value)
        print("Your final amount after applying discount code is: $", final_amount)
    else:
        print("Your final amount is : $", total_order_amount," - Discount code doesn't apply for purchases below $100!")
```

```python
# 8. Scenario: A banking system processes withdrawal requests and ensures the user has enough balance.
# write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.
```

```python
def bank_withdrawal_limit():
    user_acct_details={101:450,102:560,103:900,104:300,105:1500}
    user_cust_id=int(input("Enter customer ID:"))
    #user_withdraw_amt=int(input("Enter the withdrawl amount:"))
    min_bal=500
    if(user_cust_id in user_acct_details):
        current_balance = user_acct_details[user_cust_id]
        if(current_balance<min_bal):
            print("Withdrawl denied as there is no sufficient balance")
        else:
            user_withdraw_amt=int(input("Enter the withdrawl amount: $"))
            print("Withdraw successful. Remaining balance is : $",current_balance-user_withdraw_amt)
    else:
        print("Invalid customer ID, pls try again")
```

```python
# 9. Scenario: A calendar system verifies whether a given year is a leap year based on standard leap year rules.
# Write logic to determine whether a given year is a leap year
```

```python
def leap_year():
    v_year=int(input("Please enter a year:"))
    v_leap_calc_1=(v_year%400)
    v_leap_calc_2=(v_year%4)
    if(v_leap_calc_1==0) or  ((v_leap_calc_2==0) and ((v_year%100)!=0)):
        print(v_year, "is a leap year")
    else:
        print(v_year, "is not a leap year")
```
