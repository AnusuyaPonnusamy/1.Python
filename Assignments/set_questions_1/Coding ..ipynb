```python
'''
5.Write a Python program to check if a number is prime. Expected Output (for input 7):
7 is a prime number
'''

```

```python
def prm_num():
    num1=int(input("Enter a number:"))
    if(num1<=1):
        print(num1, "is not a prime number")
    else:
        for i in range(2, num1):
            if(num1%i==0):
                print(num1, "is not a prime number")
                break
        else:
            print(num1, "is a prime number")
```
