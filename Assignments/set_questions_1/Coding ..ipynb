```python
#A prime number is a number greater than 1 that is divisible only by 1 and itself
num1=int(input("Enter a number:"))
if(num1<=1):
    print(num1, "is not a prime number")
else:
    for i in range(2, num1+1):
        if(num1%i==0):
            print(num1, "is not a prime number")
        #else:
            #print(num1,"is a prime number")
    print(num1,"is a prime number")
```
