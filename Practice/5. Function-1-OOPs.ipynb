#FUNCTIONS: 
#1. Function is just a procedure - it is like a formula. Function gets defined in the memory.
#2. Function should always be called (or) invoked to get the output using a function call
#3. Benefit of function: Ability to return a function output using different variables in a #single line of code - Code Reusability
#4. Function name and variable name can be same
#5. Return statement should be used, when you want to pass the output of a function to another code or function
#6. Function with no parameters inside the paranthesis is called Non-Parameterized function
#7. Function with parameters inside the paranthesis is called Parametrized function and the parameters also called arguments

```python
#Function with return statement
def AgeCategory():
        if(age<18):
            print("Children")
            cate="Children"
        elif(age<35): 
            print("Adult")
            cate="Adult"
        elif(age<59):
            print("Citizen")
            cate="Citizen"
        else:
            print("Senior Citizen")
            cate="Senior Citizen"
        return cate
```

```python
age=int(input("Enter your age:"))
agecate=AgeCategory()
```

```python
print(agecate)
```

```python
#Function without return statement
def AgeCategoryNoReturn():
        if(age<18):
            print("Children")
            cate="Children"
        elif(age<35): 
            print("Adult")
            cate="Adult"
        elif(age<59):
            print("Citizen")
            cate="Citizen"
        else:
            print("Senior Citizen")
            cate="Senior Citizen"
```

```python
#Function to find odd or even numbers
def oddEven():
    num=int(input("Enter the number:"))
    if((num%2)==1):
        print("Odd Number")
        msg="Odd Number"
    else:
        print("Even Number")
        msg="Even Number"
    return msg
```

```python
oddEven()
```

```python
v_odd_even=oddEven()
```

```python
print(v_odd_even)
```

```python
#calling the v_odd_even in another condition
if(v_odd_even=="Even Number"):
    print("Hurray, you are eligible for lucky draw")
else:
    print("Sorry, please try next time")
```

```python
def BMI():
    BMI = int(input("Enter the BMI Index:"))
    if(BMI<18.5):
        print("Underweight")
        msg_1="Underweight"
    elif(BMI<24.9):
        print("Normal")
        msg_1="Normal"
    elif(BMI<29.9):
        print("Overweight")
        msg_1="Overweight"
    else:
        print("Very Overweight")
        msg_1="Very Overweight"
    return msg_1
```

```python
bmi=BMI()
```

```python
print(bmi)
```

```python
#Parameterized Function / Arguments
def addition(num1,num2):
    add=num1+num2
    return add
```
