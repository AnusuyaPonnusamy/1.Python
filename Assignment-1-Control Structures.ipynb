```python
bmi_index=float(input("Enter the BMI Index:"))
if(bmi_index<25.0):
    print("Under Weight")
elif(bmi_index>=25.0) and (bmi_index<=29.9):
    print("Increased")
elif(bmi_index>=30.0) and (bmi_index<=34.9):
    print("Very Overweight")
elif(bmi_index>=35.0) and (bmi_index<=39.9):
    print("Obese")
else:
    print("Very Obese")
```
